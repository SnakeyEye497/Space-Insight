```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, precision_recall_curve
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf

# Configuration
IMG_SIZE = 299
BATCH_SIZE = 32
MODEL_PATH = '/content/drive/MyDrive/Ml-project/model_trained/inceptionv3_model.keras'  # Update with your model path
DATA_DIR = '/content/drive/MyDrive/Ml-project/Project-Review-3 /Dataset'  # Update with your data path

def evaluate_model(model, test_generator):
    """Evaluate model and return predictions and metrics"""
    # Get predictions
    print("Generating predictions...")
    predictions = model.predict(test_generator)
    y_pred = np.argmax(predictions, axis=1)
    y_true = test_generator.classes

    # Get class names
    class_names = list(test_generator.class_indices.keys())

    # Calculate accuracy
    accuracy = np.mean(y_pred == y_true)
    print(f"Test Accuracy: {accuracy:.4f}")

    # Print classification report
    print("\nClassification Report:")
    report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)
    print(classification_report(y_true, y_pred, target_names=class_names))

    return predictions, y_true, y_pred, class_names, report

def plot_confusion_matrix(y_true, y_pred, class_names, normalize=False):
    """Plot confusion matrix with option to normalize"""
    plt.figure(figsize=(12, 10))

    if normalize:
        cm = confusion_matrix(y_true, y_pred, normalize='true')
        fmt = '.2f'
        title = 'Normalized Confusion Matrix'
    else:
        cm = confusion_matrix(y_true, y_pred)
        fmt = 'd'
        title = 'Confusion Matrix'

    sns.heatmap(cm, annot=True, fmt=fmt, cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(title)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.savefig(f"{title.lower().replace(' ', '_')}.png")
    plt.show()

def plot_roc_curves(predictions, y_true, class_names):
    """Plot ROC curves for each class (one-vs-rest)"""
    n_classes = len(class_names)

    # Compute ROC curve and ROC area for each class
    fpr = {}
    tpr = {}
    roc_auc = {}

    plt.figure(figsize=(12, 10))

    for i in range(n_classes):
        # For each class, calculate ROC using one-vs-rest approach
        fpr[i], tpr[i], _ = roc_curve((y_true == i).astype(int), predictions[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

        # Plot each class's ROC curve
        plt.plot(fpr[i], tpr[i], lw=2,
                 label=f'ROC curve of class {class_names[i]} (area = {roc_auc[i]:.2f})')

    # Plot the diagonal reference line
    plt.plot([0, 1], [0, 1], 'k--', lw=2)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Multi-class ROC Curves (One-vs-Rest)')
    plt.legend(loc="lower right")
    plt.savefig("roc_curves.png")
    plt.show()

def plot_precision_recall(predictions, y_true, class_names):
    """Plot precision-recall curves for each class"""
    n_classes = len(class_names)

    plt.figure(figsize=(12, 10))

    for i in range(n_classes):
        precision, recall, _ = precision_recall_curve((y_true == i).astype(int), predictions[:, i])
        plt.plot(recall, precision, lw=2,
                 label=f'Class: {class_names[i]}')

    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curves')
    plt.legend(loc="best")
    plt.savefig("precision_recall_curves.png")
    plt.show()

def plot_class_distribution(y_true, class_names):
    """Plot class distribution in the test set"""
    class_counts = pd.Series(y_true).value_counts().sort_index()

    plt.figure(figsize=(12, 6))
    ax = sns.barplot(x=list(range(len(class_counts))), y=class_counts.values)
    plt.xlabel('Classes')
    plt.ylabel('Number of Samples')
    plt.title('Class Distribution in Test Set')
    plt.xticks(range(len(class_names)), class_names, rotation=45, ha='right')

    # Add count labels on top of each bar
    for i, count in enumerate(class_counts.values):
        ax.text(i, count + 5, str(count), ha='center')

    plt.tight_layout()
    plt.savefig("class_distribution.png")
    plt.show()

def plot_per_class_metrics(report, class_names):
    """Plot per-class precision, recall, and F1-score"""
    # Extract metrics for each class from the classification report
    metrics = ['precision', 'recall', 'f1-score']
    df = pd.DataFrame({metric: [report[cls][metric] for cls in class_names] for metric in metrics},
                      index=class_names)

    # Plot metrics
    df.plot(kind='bar', figsize=(14, 7))
    plt.title('Per-Class Performance Metrics')
    plt.xlabel('Classes')
    plt.ylabel('Score')
    plt.ylim([0, 1.05])
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Metrics')
    plt.tight_layout()
    plt.savefig("per_class_metrics.png")
    plt.show()

def plot_misclassifications(y_true, y_pred, class_names):
    """Plot a matrix showing misclassifications between classes"""
    # Create a misclassification matrix
    n_classes = len(class_names)
    misclass_matrix = np.zeros((n_classes, n_classes))

    for i in range(len(y_true)):
        true_class = y_true[i]
        pred_class = y_pred[i]
        misclass_matrix[true_class][pred_class] += 1

    # Convert to percentages (row-wise normalization)
    row_sums = misclass_matrix.sum(axis=1, keepdims=True)
    misclass_matrix_norm = misclass_matrix / row_sums

    plt.figure(figsize=(12, 10))
    sns.heatmap(misclass_matrix_norm, annot=True, fmt='.2f', cmap='YlGnBu',
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Misclassification Matrix (Normalized)')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.savefig("misclassification_matrix.png")
    plt.show()

def main():
    # Load the trained model
    print(f"Loading model from {MODEL_PATH}...")
    model = load_model(MODEL_PATH)
    model.summary()

    # Set up data generator for test data
    test_datagen = ImageDataGenerator(rescale=1./255)

    test_generator = test_datagen.flow_from_directory(
        DATA_DIR,
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=False
    )

    # Evaluate the model
    predictions, y_true, y_pred, class_names, report = evaluate_model(model, test_generator)

    # Generate various plots
    print("\nGenerating confusion matrices...")
    plot_confusion_matrix(y_true, y_pred, class_names, normalize=False)  # Raw counts
    plot_confusion_matrix(y_true, y_pred, class_names, normalize=True)   # Normalized

    print("Generating ROC curves...")
    plot_roc_curves(predictions, y_true, class_names)

    print("Generating precision-recall curves...")
    plot_precision_recall(predictions, y_true, class_names)

    print("Generating class distribution...")
    plot_class_distribution(y_true, class_names)

    print("Generating per-class metrics...")
    plot_per_class_metrics(report, class_names)

    print("Generating misclassification matrix...")
    plot_misclassifications(y_true, y_pred, class_names)

    print("\nAll evaluation plots have been generated and saved.")

    # Additionally, save metrics to a CSV file
    metrics_df = pd.DataFrame(report).transpose()
    metrics_df.to_csv('model_performance_metrics.csv')
    print("Performance metrics saved to 'model_performance_metrics.csv'")

if __name__ == "__main__":
    main()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, precision_recall_curve
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf

# Configuration
IMG_SIZE = 224
BATCH_SIZE = 32
MODEL_PATH = '/content/drive/MyDrive/Ml-project/model_trained/Copy of vgg16_model.keras'
DATA_DIR = '/content/drive/MyDrive/Ml-project/Project-Review-3 /Dataset'

def evaluate_model(model, test_generator):
    print("Generating predictions...")
    predictions = model.predict(test_generator, verbose=1)
    y_pred = np.argmax(predictions, axis=1)
    y_true = test_generator.classes
    class_names = list(test_generator.class_indices.keys())

    accuracy = np.mean(y_pred == y_true)
    print(f"Test Accuracy: {accuracy:.4f}")
    print("\nClassification Report:")
    report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)
    print(classification_report(y_true, y_pred, target_names=class_names))

    return predictions, y_true, y_pred, class_names, report

def plot_confusion_matrix(y_true, y_pred, class_names, normalize=False, title_suffix=''):
    plt.figure(figsize=(12, 10))
    cm = confusion_matrix(y_true, y_pred, normalize='true' if normalize else None)
    fmt = '.2f' if normalize else 'd'
    title = f"{'Normalized ' if normalize else ''}Confusion Matrix {title_suffix}"

    sns.heatmap(cm, annot=True, fmt=fmt, cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(title)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.savefig(f"{title.lower().replace(' ', '_')}.png")
    plt.show()

def plot_roc_curves(predictions, y_true, class_names, title_suffix=''):
    n_classes = len(class_names)
    fpr, tpr, roc_auc = {}, {}, {}

    plt.figure(figsize=(12, 10))

    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve((y_true == i).astype(int), predictions[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
        plt.plot(fpr[i], tpr[i], lw=2,
                 label=f'Class {class_names[i]} (AUC = {roc_auc[i]:.2f})')

    plt.plot([0, 1], [0, 1], 'k--', lw=2)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Multi-class ROC Curves {title_suffix}')
    plt.legend(loc="lower right")
    plt.savefig(f"roc_curves{title_suffix.lower().replace(' ', '_')}.png")
    plt.show()

def plot_precision_recall(predictions, y_true, class_names, title_suffix=''):
    plt.figure(figsize=(12, 10))

    for i in range(len(class_names)):
        precision, recall, _ = precision_recall_curve((y_true == i).astype(int), predictions[:, i])
        plt.plot(recall, precision, lw=2, label=f'Class: {class_names[i]}')

    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title(f'Precision-Recall Curves {title_suffix}')
    plt.legend(loc="best")
    plt.savefig(f"precision_recall_curves{title_suffix.lower().replace(' ', '_')}.png")
    plt.show()

def plot_class_distribution(y_true, class_names, title_suffix=''):
    class_counts = pd.Series(y_true).value_counts().sort_index()
    plt.figure(figsize=(12, 6))
    ax = sns.barplot(x=list(range(len(class_counts))), y=class_counts.values)
    plt.xlabel('Classes')
    plt.ylabel('Number of Samples')
    plt.title(f'Class Distribution {title_suffix}')
    plt.xticks(range(len(class_names)), class_names, rotation=45, ha='right')
    for i, count in enumerate(class_counts.values):
        ax.text(i, count + 5, str(count), ha='center')
    plt.tight_layout()
    plt.savefig(f"class_distribution{title_suffix.lower().replace(' ', '_')}.png")
    plt.show()

def plot_per_class_metrics(report, class_names, title_suffix=''):
    metrics = ['precision', 'recall', 'f1-score']
    df = pd.DataFrame({metric: [report[cls][metric] for cls in class_names] for metric in metrics},
                      index=class_names)

    df.plot(kind='bar', figsize=(14, 7))
    plt.title(f'Per-Class Performance Metrics {title_suffix}')
    plt.xlabel('Classes')
    plt.ylabel('Score')
    plt.ylim([0, 1.05])
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Metrics')
    plt.tight_layout()
    plt.savefig(f"per_class_metrics{title_suffix.lower().replace(' ', '_')}.png")
    plt.show()

def plot_misclassifications(y_true, y_pred, class_names, title_suffix=''):
    n_classes = len(class_names)
    misclass_matrix = np.zeros((n_classes, n_classes))

    for i in range(len(y_true)):
        true_class = y_true[i]
        pred_class = y_pred[i]
        misclass_matrix[true_class][pred_class] += 1

    row_sums = misclass_matrix.sum(axis=1, keepdims=True)
    misclass_matrix_norm = misclass_matrix / row_sums

    plt.figure(figsize=(12, 10))
    sns.heatmap(misclass_matrix_norm, annot=True, fmt='.2f', cmap='YlGnBu',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Misclassification Matrix {title_suffix}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.savefig(f"misclassification_matrix{title_suffix.lower().replace(' ', '_')}.png")
    plt.show()

def run_full_evaluation(model, generator, name=''):
    print(f"\n---- Evaluating on {name} ----")
    predictions, y_true, y_pred, class_names, report = evaluate_model(model, generator)
    plot_confusion_matrix(y_true, y_pred, class_names, normalize=False, title_suffix=f'({name})')
    plot_confusion_matrix(y_true, y_pred, class_names, normalize=True, title_suffix=f'({name})')
    plot_roc_curves(predictions, y_true, class_names, title_suffix=f'({name})')
    plot_precision_recall(predictions, y_true, class_names, title_suffix=f'({name})')
    plot_class_distribution(y_true, class_names, title_suffix=f'({name})')
    plot_per_class_metrics(report, class_names, title_suffix=f'({name})')
    plot_misclassifications(y_true, y_pred, class_names, title_suffix=f'({name})')

    # Save report
    metrics_df = pd.DataFrame(report).transpose()
    metrics_df.to_csv(f'model_performance_metrics_{name.lower()}.csv')
    print(f"Saved report: model_performance_metrics_{name.lower()}.csv")

def main():
    print(f"Loading model from {MODEL_PATH}...")
    model = load_model(MODEL_PATH)
    model.summary()

    # Original (non-augmented) generator
    test_datagen = ImageDataGenerator(rescale=1./255)
    test_generator = test_datagen.flow_from_directory(
        DATA_DIR,
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=False
    )

    # Augmented generator (simulated test)
    aug_test_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,
        width_shift_range=0.1,
        height_shift_range=0.1,
        brightness_range=[0.8, 1.2],
        zoom_range=0.1,
        horizontal_flip=True,
        fill_mode='nearest'
    )
    aug_test_generator = aug_test_datagen.flow_from_directory(
        DATA_DIR,
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=False
    )

    # Run evaluations
    run_full_evaluation(model, test_generator, name='Original')
    run_full_evaluation(model, aug_test_generator, name='Augmented')

    print("\n✅ All evaluation plots and reports have been generated.")

if __name__ == "__main__":
    main()
```
